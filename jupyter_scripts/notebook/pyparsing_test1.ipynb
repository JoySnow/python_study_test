```python
import pyparsing as p

def erlblock_parser():
    COMMA = p.Suppress(',')
    LBRACE, RBRACE = map(p.Suppress, "{}")
    LBRACKET, RBRACKET = map(p.Suppress, "[]")
    
    z_key = p.Word(p.alphas+'_')
    z_value_tnum = p.Word(p.nums+'.')
    z_value_tword = p.Word(p.alphanums+'/"-[]:.()')
    z_value_tstr = p.OneOrMore(z_value_tword)
    z_value_tdoustrs = z_value_tstr + COMMA + z_value_tstr
    z_value_tnumstr = z_value_tnum + z_value_tstr
    z_value = z_value_tdoustrs | z_value_tnumstr | z_value_tstr | z_value_tnum
    
    z_attr = LBRACE + p.Group(z_key + COMMA + z_value) + RBRACE
    z_attr_list = p.Dict(p.ZeroOrMore(z_attr+COMMA) + p.ZeroOrMore(z_attr))
    z_block = LBRACKET + z_attr_list + RBRACKET

    z_value_plus = z_block | z_attr | z_value
    attr = LBRACE + p.Group(z_key + COMMA + z_value_plus) + RBRACE
    attr_list = p.Dict(p.OneOrMore(attr+COMMA) + p.OneOrMore(attr))
    block = LBRACKET + attr_list + RBRACKET
    
    return block


he = erlblock_parser().parseString(report10)
print type(he)
print "---he---------------------------"
print he

hed = he.asDict()
print "---hed---------------------------"
print hed

print "------------------------------"
print report10
```
